```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("work_from_home_burnout_dataset.csv")
```

```python
df = df.drop_duplicates()
```

```python
df[['work_hours','screen_time_hours','sleep_hours']].describe()
```

```python
df['burnout_risk'] = df['burnout_risk'].str.strip().str.title()
```

```python
df['productivity_index'] = df['task_completion_rate'] / df['work_hours']
```

```python
df['work_intensity'] = df['screen_time_hours'] + df['meetings_count']
```

```python
df['after_hours_work'] = df['after_hours_work'].map({0: 'No', 1: 'Yes'})
```

```python
df.info()
df.isnull().sum()
df.shape
```

```python
df.to_csv("cleaned_burnout_dataset.csv", index=False)
```

## 📘 Data Dictionary

| Column Name | Description | Data Type |
|-------------|------------|----------|
| user_id | Unique identifier for each employee | Integer |
| day_type | Indicates whether the day is a weekday or weekend | Categorical |
| work_hours | Number of hours worked in a day | Float |
| screen_time_hours | Total hours spent in front of screen | Float |
| meetings_count | Number of meetings attended in a day | Integer |
| breaks_taken | Number of breaks taken during work hours | Integer |
| after_hours_work | Indicates if employee worked after office hours (Yes/No) | Categorical |
| sleep_hours | Number of hours slept per day | Float |
| task_completion_rate | Percentage of tasks completed | Float |
| burnout_score | Calculated burnout intensity score | Float |
| burnout_risk | Category of burnout risk (Low/Medium/High) | Categorical |
| productivity_index | Derived feature: task completion rate per work hour | Float |
| work_intensity | Derived feature: screen time hours + meetings count | Float |
